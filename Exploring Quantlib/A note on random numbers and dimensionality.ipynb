```python
%matplotlib inline
import matplotlib.pyplot as plt
import QuantLib as ql

def set_unit_square(ax):
    ax.axis('scaled')
    ax.set_xlim([0,1])
    ax.set_ylim([0,1])
```

```python
rng = ql.MersenneTwisterUniformRng(42)
```

```python
xs = []
ys = []
for i in range(2047):
    xs.append(rng.next().value())
    ys.append(rng.next().value())
```

```python
    
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
set_unit_square(ax)
ax.plot(xs, ys, 'o');
```

```python
fig = plt.figure(figsize=(12,4))
for i, n in enumerate([0,1,2,3,4,5,6,7,9,11,13,15,19,23,27,31]):
    rng = ql.SoboRsg(1)
    xs = [rng.nextSequence().value()[0] for j in range(n)]
    ax = fig.add_subplot(4,4,i+1)
    ax.axis('scaled')
    ax.set_xlim([0,1])
    ax.set_ylim([-0.1,0.1])
    ax.set_xticks([])
    ax.set_yticks([])
    ax.plot(xs,[0]*len(xs))
```
